```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

**FILTRO 1- BUSQUEDA POR OBJETO**
Este filtro es utilizado para buscar puntualmente un dato dentro de alguna columna.
 En el script logramos seleccionar la variable "*NOMBRE_CLIENTE*" de la cual nos interesa ubicar las transacciones realizadas por "*MOLINERA DE MEXICO*".

```python
filtro1=df[df["NOMBRE_CLIENTE"] == "MOLINERA DE MEXICO"]
filtro1
filtro1.to_csv("Filtro1.csv")
```

**FILTRO 2- BUSCAR POR FILAS**
Utilizado para buscar especificamente alguna fila o renglón, en este caso haremos búsqueda de las filas 14, 98 y 250.

```python
filtro2=df.iloc[ [14, 98, 250] , : ]
filtro2
filtro2.to_csv("Filtro2.csv")
```

**FILTRO 3 - BUSCAR POR COLUMNAS**
Similar al filtro anterior solo que en este caso está destinado a buscar por columnas.

```python
filtro3=df.iloc[ : , [2, 4]]
filtro3
filtro3.to_csv("Filtro3.csv")
```

**FILTRO 4 - BUSQUEDA CRUZADA**
En este filtro buscamos algún dato con filas y columnas específicas.
 En el script buscamos localizar las transacciones registradas por "*ALFREDO CANELA*" y el "*MARGEN TOTAL CALCULADO*"

```python
df1= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=2)
df1
```

```python
filtro4=df1.loc[["ALFREDO CANELA"],["MARGEN_TOTAL_CALCULADO"]]
filtro4
filtro4.to_csv("Filtro4.csv")
```

**FILTRO 5 - BÚSQUEDA POR CABECERA**
Utilizado para realizar búsquedad de los elementos iniciales de un data frame.

```python
filtro5=df1.head(10)
filtro5
filtro5.to_csv("Filtro5.csv")
```

**FILTRO 6 - BÚSQUEDA POR COLA** En este fitro filtramos las filas finales del data frame

```python
filtro6=df1.tail(10)
filtro6
filtro6.to_csv("Filtro6.csv")
```

**FILTRO 7- COMPARACIÓN**
Este es utilizado para comparar datos cuantitativos siendo que, podríamos realizar operaciones aritméticas básicas.
 En este script decidimos averiguar cual es el mejor de los clientes en el registro, por lo cual procedemos a averiguar cuales de estos han gastado mas de $60,000 en las transacciones.

```python
filtro7=df[df["MARGEN_TOTAL_CALCULADO"] > 60000]
filtro7
filtro7.to_csv("Filtro7.csv")
```

**FILTRO 8 - FILTRO LÓGICO AND** 
Este es utilizado para condicionar 2 situaciones dodne ambas deben suceder simultanemanete. 
 En el script nos interesa localizar transcacciones donde se venden mas de 30 artículos y son realizadas por *ALFREDO CANELA*.

```python
filtro8=df[(df["CANT"] > 20) & (df["NOMBRE_VENDEDOR"] == "ALFREDO CANELA")]
filtro8
filtro8.to_csv("Filtro8.csv")
```

**FILTRO 9 - FILTRO LÓGICO OR** 
Este filtro lo utilizamos cuando necesitamos cumplir con características de búsqueda en específico de una u otra situación. 
 En este script buscamos información donde las transaaciones realizadas registran ventas de mas de 30 artículos o ventas realizadas por ALFREDO CANELA.

```python
filtro9=df[(df["CANT"] > 30) | (df["NOMBRE_VENDEDOR"] == "ALFREDO CANELA")]
filtro9
filtro9.to_csv("Filtro9.csv")
```

**FILTRO 10 - FILTRO LÓGICO NOT** 
Se utiliza para excluir información de búsqueda. 
 Vamos a utilizar algo de los ejemplos anteriores y buscaremos transacciones donde no se vendan menos de 20 artículos y donde ALFREDO CANELA no haya realizado la venta.

```python
filtro10=df[~(df["CANT"] < 30) & ~(df["NOMBRE_VENDEDOR"] == "ALFREDO CANELA")]
filtro10
filtro10.to_csv("Filtro10.csv")
```
